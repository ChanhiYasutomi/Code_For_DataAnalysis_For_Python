```python
#これをDataFrameの中の角括弧の中に入れればデータフレーム型になる
#西暦に2015だけ抽出される
df[df['西暦（年）'] == 2015]
```

```python
# 条件に一致する行のみデータ抽出（文字列）
# 東京都を指定して、データを抽出

df[df['都道府県名'] == '東京都'] #都道府県名が東京都の行が抽出
```

```python
# 条件に一致する行のみデータ抽出（10の倍数）
# 西暦が10の倍数のデータを抽出(10の倍数つまり、西暦を10で割った時の余りが0となる場合)
df[df['西暦（年）'] % 10 == 0]
```

```python
# 条件に一致しない行のみデータ抽出（Not演算子）
# ~ - not演算子
df[~(df['西暦（年）'] % 10 == 0)] #西暦が10の倍数ではない条件の記述
```

```python
# 条件に一致しない行のみデータ抽出（不等価演算子）
# != - 不当価演算子
df[df['西暦（年）'] % 10 != 0]
```

```python
# and条件
df[(df['西暦（年）'] == 2015) & (df['都道府県名'] == '東京都')] #西暦2015年、かつ、東京都の行を抽出
```

```python
# or条件
df[(df['西暦（年）'] == 2010) | (df['西暦（年）'] == 2015)] #西暦2010年または、2015年の行を抽出
```

```python
# 特定の条件でデータ抽出するqueryメソッド
# カラム名を変更 - '西暦（年）'から 'year'
df = df.rename(columns={'西暦（年）': 'year'})
df
```

```python
# queryメソッドを使って
df.query('year == 2010') #2010年のデータのみ抽出
```

```python
df.query('year > 2010') #2010年を超えるデータのみ抽出
```

```python
# シリーズで使えるメソッド = mapとapplyの2つ
# いずれもそれぞれの要素に対して、関数を適用させたりすることができる。

# 使い分け
# 1,ラムダ式を使って2倍にすること
# 2,if文つきのラムダ式
# 3,二乗するといったnumpy の数学関数を使うことはどちらもできる

# 文字列操作は、mapメソッドが向いている
# 関連付けをするマッピングができるのはmapメソッド

# 引数があるような関数を適用させたい場合はapplyメソッドが向いている

# 文字列を扱うときはmapメソッド
# 数字を扱うときはapplyメソッド



# データフレームで使えるメソッド = applymapとapply
# applymapメソッドは、それぞれの要素に対して関数を適用させる

# 一方のapplyメソッドは、適用させたい関数の種類によって、それぞれの要素に対してか、列や行ごとへの適用ができる
# 1,ラムダ式で2倍にしたり
# 2,numpy数学関数を使う場合はどちらでもできた

# applymapメソッド - if文付きの関数や文字列操作に適している
# applyメソッド - 集計をしたい場合や引数がある関数を適用させたい場合



# DataFrame
# 引数がない関数,ラムダ式 applymap:O, apply:O
# if文付きラムダ式,関数 applymap:O, apply:x
# 文字列操作を含む関数 applymap:O, apply:x
# Numpy関数(数学関数) applymap:O, apply:O
# Numpy関数(集計関数) applymap:x, apply:O
# 引数のある関数 applymap:x, apply:O
# 関連付け(マッピング) applymap:-, apply:-

# Series
# 引数がない関数,ラムダ式 map:O, apply:O
# if文付きラムダ式,関数 map:O, apply:x
# 文字列操作を含む関数 map:O, apply:△
# Numpy関数(数学関数) map:O, apply:O
# Numpy関数(集計関数) map:x, apply:x
# 引数のある関数 map:x, apply:O
# 関連付け(マッピング) map:o, apply:x
```
